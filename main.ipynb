```python
import matplotlib.pyplot as plt
import csv
import numpy as np
from tqdm import tqdm

from trees.DecisionTree import DecisionTree as DecisionTreeRoot
from trees.RandomForest import RandomForest
```

```python
dataset = []
with open('datasets/test_2d.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    categories = next(reader)
    for row in reader:
        dataset.append(list(map(float, row)))
```

```python
data = np.array(dataset)

x = data[:, 0]
y = data[:, 1]
target = data[:, 2]

# Plotting the data points
plt.scatter(x[target == 0], y[target == 0], color='blue', label='Target 0')
plt.scatter(x[target == 1], y[target == 1], color='orange', label='Target 1')

# Adding labels and legend
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
root = DecisionTreeRoot(data, use_entropy=False)
root.generate_tree()
```

```python
root.predict([50, 50])
```

```python
root.draw(areas=True, dividing_lines=False)
```

```python
random_forest = RandomForest(data, use_all_features=True)
random_forest.predict(np.array([-10, -10]))
```
